```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
base = 'https://www.falabella.com'
r = requests.get(base+'/falabella-cl/category/cat1320008/Ver-Todo-Moda-Hombre')
soup = BeautifulSoup(r.text,'lxml')
searchResults = soup.find_all("div", {"id" : re.compile('testId-pod-\d')})
```

```python
product = []

justNumbers = re.compile(r'\d')
for n,result in enumerate(searchResults):
    product.append({'id' : result.get('id')})
    #Titles
    re.compile(r'pod-(sub)?title')
    titles = result.find_all('b',{'class' : re.compile('pod-(subTitle|title)')})
    if titles is not None:
        product[n]['Brand'] = titles[0].text
        product[n]['Type'] = titles[1].text    
        
    #Prices
    pricesDiv = result.find("div", {"id" : re.compile('testId-pod-prices.*')})
    if pricesDiv is not None:
        prices = pricesDiv.find_all('li')
        product[n]['NormalPrice'] = filter(justNumbers.search,prices[1].text)
        #CMR Credit card
        product[n]['CMRPrice'] = filter(justNumbers.search,prices[0].text)        
        
```
